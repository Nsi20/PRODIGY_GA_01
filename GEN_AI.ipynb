# **Step 2: Loading the GPT-2 Model and Tokenizer**

**2.1 Importing Necessary Libraries**

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

**2.2 Initializng Tokenizer**

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
```

**2.3 Loading the GPT-2 Model**

```python
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

**2.4 Adjusting the Tokenizer and Model for Padding**

```python
tokenizer.add_special_tokens({'pad_token': '<pad>'})
model.resize_token_embeddings(len(tokenizer))
```

**2.5 Moving the Model to GPU**

```python
model = model.to('cuda')
```

# **Step 3: Preparing a Dataset**

**3.1 Importing Dataset Libraries**

```python
!pip install datasets
```

```python
from datasets import load_dataset
```

**3.2 Loading the Wikitext-2 Dataset**

```python
dataset = load_dataset("wikitext", "wikitext-2-raw-v1")
```

**3.3 Exploring the Dataset**

```python
print(dataset)
print(dataset['train'][0])  # Print the first training example
```

**3.4 Tokenizing the Dataset**

```python
def tokenize_function(examples):
    return tokenizer(examples['text'], truncation=True, padding=True, max_length=512)

tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

**3.5 Preparing the Dataset for Training**

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)

tokenized_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask'])
```

**3.6 Spliting the Dataset**

```python
train_dataset = tokenized_dataset["train"]
validation_dataset = tokenized_dataset["validation"]
```

# **Step 4: Fine-Tune GPT-2**

**4.1 Import Training Tools**

```python
from transformers import AdamW, Trainer, TrainingArguments
```

**4.2 Set Training Arguments**

```python
training_args = TrainingArguments(
    output_dir="./results",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    save_steps=500,
    evaluation_strategy="steps",
    save_total_limit=2,
    fp16=True,
)
```

**4.3 Initializing Trainer**

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
)
```

# **Step 5 : Loading the Model and Generate Text**

**5.1 Loading the Model and Tokenizer**

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

model = GPT2LMHeadModel.from_pretrained("./fine_tuned_gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("./fine_tuned_gpt2")

model = model.to('cuda')
```

**5.2 Generating Text from a Prompt: Defining a prompt and generating text:**

```python
prompt = "Once upon a time in a distant galaxy,"
input_ids = tokenizer.encode(prompt, return_tensors="pt").to('cuda')

output = model.generate(
    input_ids,
    max_length=50,
    num_beams=5,
    no_repeat_ngram_size=2,
    early_stopping=True
)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load model and tokenizer
model = GPT2LMHeadModel.from_pretrained("./fine_tuned_gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("./fine_tuned_gpt2")

# Set device (GPU or CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Set pad token
tokenizer.pad_token = tokenizer.eos_token

# Prepare input
prompt = "Once upon a time in a distant galaxy,"
input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)
attention_mask = (input_ids != tokenizer.pad_token_id).to(device)

# Generate text
output = model.generate(
    input_ids=input_ids,
    attention_mask=attention_mask,
    max_length=100,
    temperature=0.7,
    top_p=0.9,
    top_k=50,
    repetition_penalty=1.2,
    do_sample=True,
    pad_token_id=tokenizer.eos_token_id  # Set explicitly
)

# Decode and print
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```
